```python
# Dependencies
from bs4 import BeautifulSoup
import requests
from splinter import Browser
import pymongo
import pandas as pd
```

```python
# Set URL's
news_url= "https://mars.nasa.gov/news/"
img_url= "https://www.jpl.nasa.gov/spaceimages/?search=&category=Mars"
fact_url= "https://space-facts.com/mars/"
usgs_url= "https://astrogeology.usgs.gov/search/results?q=hemisphere+enhanced&k1=target&v1=Mars"
```

```python
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
browser.visit(news_url)

html = browser.html

soup = BeautifulSoup(html, 'html.parser')
```

```python
# Create a variable to store the scraped article title and paragraph description
first_article = soup.find_all('div', class_='content_title')[1].text
first_para = soup.find_all('div', class_='article_teaser_body')[0].text
print(first_article)
print(first_para)
```

```python
# Retrieve page with the browser module
browser.visit(img_url)
page = requests.get(img_url)
soup = BeautifulSoup(page.text, 'html.parser')
print(soup.prettify())
```
